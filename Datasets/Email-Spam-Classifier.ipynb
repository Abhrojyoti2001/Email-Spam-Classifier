```python
path = '/kaggle/input/ham-and-spam-dataset/hamnspam/'
emails = []
labels = []
for label in ['ham/','spam/']:
    filenames = os.listdir(os.path.join(path,label))
    for file in filenames:
        f = open(os.path.join(path,label,file),'r',encoding='latin-1')
        bolk = f.read()
        emails.append(bolk)
        labels.append(label)
```

```python
df = pd.DataFrame({'emails':emails,'labels':labels})
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['labels'] = encoder.fit_transform(df['labels'])
```

```python
df.head()
```

```python
# text prepocessing
# missing data
# duplicate
# lower case
# special chars
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['emails'] = df['emails'].apply(lambda x:x.lower())
```

```python
df['emails'] = df['emails'].apply(lambda x:x.replace('\n', " "))
df['emails'] = df['emails'].apply(lambda x:x.replace('\t', " "))
```

```python
def remove_special_char(text):
    new_text = ""
    for i in text.split():
        if i.isalpha():
            new_text = new_text + " " + i
        
    return new_text
```

```python
df['emails'] = df['emails'].apply(remove_special_char)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=3000,stop_words='english')
```

```python
X = cv.fit_transform(df['emails']).toarray()
```

```python
X.shape
```

```python
y = df['labels'].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test, = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB,MultinomialNB,BernoulliNB,ComplementNB
gnb = GaussianNB()
mnb = MultinomialNB()
bnb = BernoulliNB()
cnb = ComplementNB()
```

```python
gnb.fit(X_train,y_train)
mnb.fit(X_train,y_train)
bnb.fit(X_train,y_train)
cnb.fit(X_train,y_train)
```

```python
y_pred1 = gnb.predict(X_test)
y_pred2 = mnb.predict(X_test)
y_pred3 = bnb.predict(X_test)
y_pred4 = cnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred1))
print(accuracy_score(y_test,y_pred2))
print(accuracy_score(y_test,y_pred3))
print(accuracy_score(y_test,y_pred4))
```

```python
import pickle
pickle.dump(cv,open('vectors.pkl','wb'))
pickle.dump(cnb,open('model.pkl','wb'))
```
